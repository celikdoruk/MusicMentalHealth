### Loading the library and the csv file into a dataframe

```python
import pandas as pd
```

```python
df = pd.read_csv('mxmh_survey_results.csv')
```

```python
df
```

### Creating a unique submission id column, to later on use as a PRIMARY KEY in PostgreSQL 

```python
df['submission_id'] = [i for i in range(1, len(df) + 1)]
```

```python
submission_id_col = ['submission_id']
rest_of_the_cols = [col for col in df.columns if col not in submission_id_col]
```

```python
# rearranging the dataframe
df = df[submission_id_col + rest_of_the_cols]
```

```python
df
```

### Dividing the dataset into 3 parts. These tables will later be joined using SQL.

```python
user_info_df = df.loc[:, 'submission_id': 'BPM']
```

```python
frequency_df = df.loc[:, 'Frequency [Classical]': 'Frequency [Video game music]']
```

```python
mental_health_df = df.loc[:, 'Anxiety': 'Music effects']
```

### Adding the unique 'submission_id' column to the new dataframes.

```python
frequency_df['submission_id'] = [i for i in range(1, len(frequency_df) + 1)]
submission_id_col = ['submission_id']
rest_of_the_cols = [col for col in frequency_df.columns if col not in submission_id_col]
frequency_df = frequency_df[submission_id_col + rest_of_the_cols]
```

```python
mental_health_df['submission_id'] = [i for i in range(1, len(mental_health_df) + 1)]
submission_id_col = ['submission_id']
rest_of_the_cols = [col for col in mental_health_df.columns if col not in submission_id_col]
mental_health_df = mental_health_df[submission_id_col + rest_of_the_cols]
```

### Taking a look at the datatypes of our tables

### Dumping the dataframes as .csv files

```python
user_info_df.to_csv('user_information.csv', index=None)
frequency_df.to_csv('frequency.csv', index=None)
mental_health_df.to_csv('menal_health_scores.csv', index=None)
```
